#### Performing Calculations on ints and floats

##### Addition

```python
num1 = 2
num2 = 5.4

result = num1 + num2

print(result)

# Short-hand version
result += 2
print(result)
```

##### Subtraction

```python
result_diff = num2 - num1
print(result_diff)

# Short-hand version
result_diff-= 1
print(result_diff)
```

##### Multiplication

```python
result_mul = num1 * num2
print(result_mul)

# Short-hand version
result_mul *= 2
print(result_mul)
```

##### Division

```python
result_div = num2/num1
print(result_div)

# Short-hand result
result_div /= 2
print(result_div)
```

##### Floor Division

```python
result_floor = num2 // num1
print(result_floor)

# Short-hand version
result_floor //= 2
print(result_floor)
```

##### Modulo

```python
result_mod = num2 % num1
print(result_mod)
```

##### Exponential

```python
square = 5 ** 2
print(square)

# Short-hand version
square **= 2
print(square)
```

### If Statements

```python
# Available operators: Greater(>), Less(<),Equal(==)
# Greater or Equal(>=), Less or Equal (<=)

# Truth Tree:
# T && F = F
# T && T = T
# T || F = T
# F || T = T
# F || F = F

num1 = 5
num2 = 10

if num2 == num1:
    print("Equal Values")
else:
    print("Not Equal Values")
    
if num2 >= num1:
    print("Num2 is great")
elif num1 <= num2:
    print("Num1 is less")
```

### Lists

##### Declaring Lists

```python
l_1 = [] # empty list declaration

names = ['Max', 'Cindy', 'Kattie', 'Bob', 'Nate']
print(names)
```

##### Indexing a List

```python
# [start: stop: step] -- [-1]

# single index
print(names[0])

# print starting at index 1 going to end
print(names[1:])

# print starting at the begining of the list to n (n == the number we want to start)
print(names[:2])

# print starting index 1 going up by 2 in each iteration
print(names[1::2])

# print starting at the back and present reverse order
print(names[::-1])
```

##### .append()

```python
names.append("Josh")
print(names)
```

##### .insert()

```python
# list.insert(index, value)

names.insert(3, "Gabby")
print(names)
```

##### .pop()

```python
# default is the last value if no parameter is given
# pop returns the element that was removed in case you want to assign it into a variable

my_name = names.pop()
print(my_name)

print(names.pop(3))
```


##### .remove()

```python
# value to be removed, not an index

# names.remove('Bob')

while "Bob" in names:
    names.remove('Bob')
print(names)
```

##### del()

```python
# index to be removed, not value

#Be Careful with del, can cause an indexing error if not used carefully

del(names[1])
print(names)
```

##### Concatenating Two Lists

```python
# will append two lists together, not add up the elements in each

l_2 = [0, 1, 2]
l_3 = [3, 4, 5]

large_list = l_2 + l_3
print(large_list)
```

##### Lists Within Lists

```python
# Lists are able to hold any type of element, including other lists
# They can also go as deep as you want, having many lists nested inside of eachother

names = ['Nate', 'Ryan', 'Gaby', ['Joel', 'Terrell', 'Derek']]
print(names)

print(names[3][2])
```

##### Looping Through Lists

```python
# Two ways to loop through a list, one is by index, the other is by using the 'in' keyword

# loop with a for loop(by index)
for i in range(len(names)):
    print(names[i])
    
# loop with a for loop using 'in' keyword
for i in names:
    print(i)
```

## Exercise #1 <br>
<p>Cube Number Test... Print out all cubed numbers up to the total value 1000. Meaning that if the cubed number is over 1000 break the loop.</p>

```python
#Simple Method
for i in range(1, 11):
    print(i ** 3)
    
# extra
for i in range(11):
    if i ** 3 > 1000:
        continue
    print(i ** 3)
```

## Exercise #2 <br>
<p>Get first prime numbers up to 100</p>

```python
# HINT::
# An else after an if runs if the if didn’t
# An else after a for runs if the for didn’t break

for n in range(2, 101):
    if n > 1:
        for i in range(2,n):
            if n % i == 0:
                break
        else:
            print(n)

num = 2

while num < 101:
    for i in range(2,num):
        if num % i == 0:
            break
    else:
        print(num)
    num+=1


            
            
    
    

     
```
